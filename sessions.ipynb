```python
import pandas as pd
from matplotlib import pyplot as plt
import prophet as pf
```

```python
data = pd.read_csv("data/HistoricalData_1759454414777.csv")
```

```python
data.head()
```

```python
data['Open'] = data['Open'].str.replace('$','').astype(float)
data['Close/Last'] = data['Close/Last'].str.replace('$','').astype(float)
data['High'] = data['High'].str.replace('$','').astype(float)
data['Low'] = data['Low'].str.replace('$','').astype(float)
```

```python
data.plot(x='Date', y=['Open'], kind='line', title='Stock Prices Over Time', figsize=(10, 6))
data.plot(x='Date', y=['Volume'], kind='line', title='Stock Prices Over Time', figsize=(10, 6))
```
